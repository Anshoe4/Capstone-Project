```python
import pandas as pd
import re

# Step 1: Define a function to create a standardized match key
def extract_match_key(company_name):
    if pd.isna(company_name):
        return ""
    # Convert to lowercase
    company_name = company_name.lower()

    # Replace "pvt." (or "pvt") with "private" and "ltd." (or "ltd") with "limited"
    company_name = re.sub(r"\bpvt\.?\b", "private", company_name, flags=re.IGNORECASE)
    company_name = re.sub(r"\bltd\.?\b", "limited", company_name, flags=re.IGNORECASE)

    # Remove anything within brackets (non-greedy)
    company_name = re.sub(r"\s*\(.*?\)\s*", "", company_name)

    # Remove all non-alphanumeric characters
    company_name = re.sub(r"[^a-z0-9]", "", company_name)

    return company_name

# Assume df_excel and df are your two DataFrames
# Step 2: Apply the cleaning function to both dataframes
df_excel['Cleaned_Name'] = df_excel['Company'].apply(extract_match_key)
df['Cleaned_Name'] = df['sa_company_name'].apply(extract_match_key)
```

```python
import pandas as pd
import re
from google.colab import files

# Function to clean company names
def clean_company_name(name):
    if pd.isna(name):
        return ""
    name = name.lower()
    name = re.sub(r'\b(pvt\.?|ltd\.?|private|limited)\b', '', name)
    name = re.sub(r'\(.*?\)', '', name)  # Remove anything inside brackets
    name = re.sub(r'[^a-zA-Z0-9]', '', name)  # Remove all special characters
    return name.strip()

# Apply cleaning function to both datasets
df['Cleaned_Name'] = df['sa_company_name'].apply(clean_company_name)
df_excel['Cleaned_Name'] = df_excel['Company'].apply(clean_company_name)

# Merge with indicator to check matching
merged_check = df_excel.merge(df, on="Cleaned_Name", how="left", indicator=True)

# Extract unmatched companies from df_excel
unmatched_companies = merged_check[merged_check['_merge'] == 'left_only'][['Company']]
unmatched_companies.to_csv("unmatched_companies.csv", index=False)

# Perform full join and add PE_transaction & Transaction_Date columns
df_merged = df.merge(df_excel[['Cleaned_Name', 'Transaction Date']], on="Cleaned_Name", how="left")
df_merged['PE_transaction'] = df_merged['Transaction Date'].notna().astype(int)  # 1 if matched, else 0

# Save the merged dataset with PE transaction info
df_merged.to_csv("df_with_PE_transaction.csv", index=False)

# Download both files
files.download("unmatched_companies.csv")
files.download("df_with_PE_transaction.csv")
```
